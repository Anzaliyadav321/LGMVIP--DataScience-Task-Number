# Task-4 Image to Pencil Sketch With Python (Beginner - Level)

# Importing Library

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

# Reading Image

```python
image1 = cv2.imread("dog pic2.jpg")
cv2.imshow("Dog", image1)
cv2.waitKey(0)
```

```python
plt.imshow(image1)
plt.axis(False)
plt.show()
```

```python
image2 = cv2.imread("bird pic1.jpg")
cv2.imshow("Bird", image2)
cv2.waitKey(0)
```

```python
plt.imshow(image2)
plt.axis(False)
plt.show()
```

# Converting image to gray scale

```python
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
cv2.imshow("New Dog", gray_image1)
cv2.waitKey(0)
```

```python
plt.imshow(gray_image1)
plt.axis(False)
plt.show()
```

```python
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
cv2.imshow("New Bird", gray_image2)
cv2.waitKey(0)
```

```python
plt.imshow(gray_image2)
plt.axis(False)
plt.show()
```

# Inverting the new grayscale image

```python
inverted_image1 = 255 - gray_image1
cv2.imshow("Inverted1", inverted_image1)
cv2.waitKey()
```

```python
plt.imshow(inverted_image1)
plt.axis(False)
plt.show()
```

```python
inverted_image2 = 255 - gray_image2
cv2.imshow("Inverted2", inverted_image2)
cv2.waitKey()
```

```python
plt.imshow(inverted_image2)
plt.axis(False)
plt.show()
```

# Blur the image by using the Gaussian Function in OpenCV

```python
blurred1 = cv2.GaussianBlur(inverted_image1, (21,21), 0)
```

```python
blurred2 = cv2.GaussianBlur(inverted_image2, (21,21), 0)
```

# Inverting the blurred image, then we can easily convert the image into a pencil sketch

```python
inverted_blurred1 = 255 - blurred1
pencil_sketch1 = cv2.divide(gray_image1, inverted_blurred1, scale=256.0)
cv2.imshow("Sketch1", pencil_sketch1)
cv2.waitKey(0)
```

```python
plt.imshow(pencil_sketch1)
plt.axis(False)
plt.show()
```

```python
inverted_blurred2 = 255 - blurred2
pencil_sketch2 = cv2.divide(gray_image2, inverted_blurred2, scale=256.0)
cv2.imshow("Sketch2", pencil_sketch2)
cv2.waitKey(0)
```

```python
plt.imshow(pencil_sketch2)
plt.axis(False)
plt.show()
```

# looking at both the original image and the pencil sketch

```python
cv2.imshow("original image1", image1)
cv2.imshow("pencil sketch1", pencil_sketch1)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image1', size=18)
plt.imshow(image1)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('pencil sketch1', size=18)
rgb_sketch=cv2.cvtColor(pencil_sketch1, cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch1)
plt.axis('off')
plt.show()
```

```python
cv2.imshow("original image2", image2)
cv2.imshow("pencil sketch2", pencil_sketch2)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image2', size=18)
plt.imshow(image2)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('pencil sketch2', size=18)
rgb_sketch=cv2.cvtColor(pencil_sketch2, cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch2)
plt.axis('off')
plt.show()
```
